```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
urls = [
    "https://raw.githubusercontent.com/delinai/schulich_ds1_2024/main/Datasets/customers_final.csv",
    "https://raw.githubusercontent.com/delinai/schulich_ds1_2024/main/Datasets/engagements_final.csv",
    "https://raw.githubusercontent.com/delinai/schulich_ds1_2024/main/Datasets/marketing_final.csv",
    "https://raw.githubusercontent.com/delinai/schulich_ds1_2024/main/Datasets/transactions_final.csv"
]

dataframes = []

# Loop through each URL and read the CSV file into a DataFrame
for url in urls:
    df = pd.read_csv(url)
    dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
df = pd.concat(dataframes, ignore_index=True)

# Display the combined DataFrame
print(df)
```
